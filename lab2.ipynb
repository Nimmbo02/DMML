```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import sklearn.metrics as sm
from scipy import stats
```

```python
from scipy import stats
def remove_outliner(df, threshold):
    z_scores = np.abs(stats.zscore(ford)) # Adjust the threshold as needed
    outlier_indices = np.where(z_scores > threshold)
    #remove outliers
    for i in outlier_indices:
        ford.drop(ford.index[i], inplace=True)
```

```python
# spliting the data into train and test
train_set, test_set = train_test_split(ford, test_size=0.2, random_state=40)
# dropping the price column from the train set
X_train = train_set.drop("price", axis=1)
# taking the price column from the train set
y_train = train_set["price"]
```

```python
# scaling the data
std_scaler = StandardScaler()
X_train = std_scaler.fit_transform(X_train)
```

```python
# creating the linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
#remove the price column from the test set
X_test = test_set.drop("price", axis=1)
y_test = test_set["price"]
std_scaler = StandardScaler()
X_test = std_scaler.fit_transform(X_test)
```

```python
# get the predictions for the test set
y_pred = lin_reg.predict(X_test)
```

```python
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2)) 
print("R2 score =", round(sm.r2_score(y_test, y_pred), 2))
```
